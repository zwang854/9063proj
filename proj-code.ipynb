```python

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from math import sqrt
```

```python
#extract time series data
def extract_data(data,slide):
    X=[]
    y=[]
    for i in range(len(data)-slide):
        X.append([a for a in data[i:i+slide]])
        y.append(data[i+slide])
    X=np.array(X)
    X=X.reshape(X.shape[0],X.shape[1],1)
    return X,y


```

```python
acc_num=date_df.loc[:,'acc#']
train, test = acc_num[0:-(365*4)], acc_num[-(365*4):]
print(max(train))
```

```python
train_norm=train/max(train)

train_norm
```

```python
fig1=plt.figure(figsize=(15,5))
plt.plot(train)
```

```python
time_step=7
```

```python
X,y=extract_data(train_norm,time_step)
#print(X[0,:,:])
#print(y)
```

```python
#set up model
from keras.models import Sequential
from keras.layers import Dense,SimpleRNN
```

```python
model=Sequential()
model.add(SimpleRNN(units=5,input_shape=(time_step,1),activation='relu'))
model.add(Dense(units=1,activation='linear'))
model.compile(optimizer='adam',loss='mean_squared_error')
model.summary()
```

```python
print(X.shape,len(y))
```

```python
X = np.asarray(X)
y = np.asarray(y)
#validation_x = np.asarray(validation_x)
#validation_y = np.asarray(validation_y)
model.fit(X,y,batch_size=60,epochs=100)
```

```python
import math
```

```python
y_train_predict = model.predict(X)*max(train)
y_train = y*max(train)
y_train_predict = [math.floor(i[0]) for i in y_train_predict]
```

```python
fig2=plt.figure(figsize=(15,5))
plt.plot(y_train,label='real')
plt.plot(y_train_predict,label='predict')
plt.title('Number of Accidents on Daily Base(train)')
plt.xlabel('time')
plt.ylabel('#acc')
plt.show()
```

```python
test_norm=test/max(train)

X_test_norm,y_test_norm=extract_data(test_norm,time_step)
print(X_test_norm.shape, len(y_test_norm))
```
