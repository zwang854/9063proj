```python
#extract time series data
def extract_data(data,slide):
    X=[]
    y=[]
    for i in range(len(data)-slide):
        X.append([a for a in data[i:i+slide]])
        y.append(data[i+slide])
    X=np.array(X)
    X=X.reshape(X.shape[0],X.shape[1],1)
    return X,y


```
